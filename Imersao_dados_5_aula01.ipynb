```python
rua_sem_nome = dados[dados['Rua'].isna()]
rua_sem_nome.groupby(['Bairro']).agg(count = ('Bairro','count')).sort_values(by='count', ascending=False).head(20).plot.bar()
```

```python
rua_sem_nome.groupby(['Bairro']).agg(contagem = ('Bairro','count')).sort_values(by='contagem', ascending=False)
```

```python
dados.groupby(['Bairro']).agg( 
    mean_quartos = ('Quartos','mean'),
    mean_banheiros = ('Banheiros','mean')).sort_values(by='mean_quartos', ascending=False).plot(secondary_y = 'mean_quartos')
```

```python
dados.query(("Metragem > 3000"))
```

```python
dados.query(("Metragem < 2000")).groupby(['Bairro']).agg( 
    mean_quartos = ('Quartos','mean'),
    mean_banheiros = ('Banheiros','mean'),
    mean_metragem = ('Metragem','mean')).plot.scatter(x='mean_banheiros',
                                                      y='mean_quartos',
                                                      c='mean_metragem',
                                                      colormap='plasma')
```
