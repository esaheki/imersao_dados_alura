```python
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

```python
dados[["Moeda", "Valor_anuncio", "Tipo_anuncio"]] = dados["Valor"].str.split(expand = True)
dados.head(50)
```

```python
dados[dados["Tipo_anuncio"].isnull()]["Tipo_anuncio"].unique()
```

```python
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_anuncio"].str.replace(".","", regex=False).astype(float)
```

```python
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".","", regex=False).astype(float)
```

```python
dados_vendas.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10, 8))
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_float", kde = True)
ax.set_title("Histograma Valor Imóvel")
plt.xlim((-50, 10000000))
plt.show()
```

```python
sns.set()
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
plt.figure(figsize=(15, 8))
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_milhoes", kde = False, color = "teal", lw=2, bins=65, edgecolor="white")
ax.set_title("Imóveis em São Paulo")
ax.set(xlabel='Valor do imóvel em milhões de reais', ylabel='Quantidade de imóveis à venda')
plt.xlim((-0.5, 10.5))
plt.show()
```

```python
dados_vendas["Valor_metro_quadrado"] = (dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"])
```

Explorando os valores, tipos de graficos e bibliotecas

```python
sns.jointplot(data=dados_vendas, x="Metragem", y="Valor_anuncio_float", height=10)
```

```python
sns.jointplot(data=dados_vendas.query("Metragem < 10000"), x="Metragem", y="Valor_anuncio_float", height=10)
```

```python
plot_data = dados_vendas.groupby(['Bairro']).agg( 
    mean_metragem = ('Metragem','mean'),
    mean_valor_metro = ('Valor_metro_quadrado','mean'),
    count = ('Bairro', 'count'))


plt.figure(figsize=(20, 10))
sns.scatterplot(
    data=plot_data, x="mean_metragem", y="mean_valor_metro", hue="count", size="count", palette="viridis", sizes=(10, 800)
)
```

```python
plot_data = dados_vendas.query("Metragem < 10000 & Valor_metro_quadrado < 80000").groupby(['Bairro']).agg( 
    mean_metragem = ('Metragem','mean'),
    mean_valor_metro = ('Valor_metro_quadrado','mean'),
    count = ('Bairro', 'count'))


plt.figure(figsize=(20, 10))
sns.scatterplot(
    data=plot_data, x="mean_metragem", y="mean_valor_metro", hue="count", size="count", palette="viridis", sizes=(10, 800)
)
```

```python
plot_data = dados_vendas.query("Metragem < 10000 & Valor_metro_quadrado < 80000").groupby("Bairro", as_index=False).mean().sort_values(by='Valor_metro_quadrado', ascending=False)
plt.figure(figsize=(20, 160))
ax = sns.barplot(data=plot_data, x="Valor_metro_quadrado", y="Bairro")
ax.set_title("Média do metro quadrado por bairros de São Paulo")
ax.set(xlabel='Valor do metro quadrado em reais', ylabel='Bairro')
plt.show()
```

```python
plot_data = dados_vendas.query("Metragem < 10000 & Valor_metro_quadrado < 80000").groupby("Bairro", as_index=False).mean().sort_values(by='Valor_metro_quadrado', ascending=False)
divisor = pd.DataFrame({'Bairro':['...']})
plt.figure(figsize=(20, 10))
ax = sns.barplot(data=pd.concat([plot_data.head(10), divisor, plot_data.tail(10)]), x="Valor_metro_quadrado", y="Bairro")
ax.set_title("Top 10 bairros com o metro quadrado mais caros e mais baratos em São Paulo", size=16)
ax.set(xlabel='Valor do metro quadrado em reais')
plt.show()
```
